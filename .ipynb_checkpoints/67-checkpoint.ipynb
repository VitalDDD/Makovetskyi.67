```python
import numpy as np
import tensorflow as tf
import matplotlib.pyplot as plt

# Створення випадкових даних
num_samples = 100
x = np.linspace(-10, 10, num_samples)
y = 2 * x + 3 + np.random.randn(num_samples) * 3  # Лінійна функція з шумом

# Побудова моделі лінійної регресії з використанням TensorFlow
model = tf.keras.Sequential([
    tf.keras.layers.Dense(units=1, input_shape=[1])  # Один нейрон вхідного шару
])

# Компіляція моделі
model.compile(optimizer='sgd', loss='mean_squared_error')

# Навчання моделі
model.fit(x, y, epochs=100)

# Оцінка моделі на датасеті
loss = model.evaluate(x, y)
print("Помилка (MSE):", loss)

# Предикція на нових даних
new_x = np.linspace(-15, 15, 20)
predictions = model.predict(new_x)

# Виведення результатів
print("Предикції:")
for i in range(len(new_x)):
    print(f"x = {new_x[i]:.2f}, y_pred = {predictions[i][0]:.2f}")

# Графік результатів
plt.scatter(x, y, label='Дані')
plt.plot(new_x, predictions, color='red', label='Прогноз')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Лінійна регресія')
plt.legend()
plt.show()
```
